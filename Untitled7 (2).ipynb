```python
import pickle
import pandas as pd
import seaborn as sns
from os.path import exists
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
file_path = '/content/drive/My Drive/CS506/train.csv'

trainingSet = pd.read_csv("/content/drive/My Drive/CS506/train.csv")
testingSet = pd.read_csv("/content/drive/My Drive/CS506/test.csv")

print("train.csv shape is ", trainingSet.shape)
print("test.csv shape is ", testingSet.shape)

print()

print(trainingSet.head())
print()
print(testingSet.head())

print()

print(trainingSet.describe())

print("Missing values in training set:")
print(trainingSet.isnull().sum())
print()
print("Missing values in testing set:")
print(testingSet.isnull().sum())
print()

trainingSet['Score'].value_counts().plot(kind='bar', legend=True, alpha=.5)
plt.title('Score Distribution (Bar Plot)')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()

print()
print("EVERYTHING IS PROPERLY SET UP! YOU ARE READY TO START")

import pandas as pd
from os.path import exists
from sklearn.model_selection import train_test_split

def add_features_to(df):
    df['Helpfulness'] = df.apply(
        lambda row: row['HelpfulnessNumerator'] / row['HelpfulnessDenominator'] if row['HelpfulnessDenominator'] != 0 else 0,
        axis=1
    )
    return df

# Define paths
train_path = "/content/drive/My Drive/CS506/X_train.csv"
submission_path = "/content/drive/My Drive/CS506/X_submission.csv"

if exists(train_path) and exists(submission_path):
    X_train = pd.read_csv(train_path)
    X_submission = pd.read_csv(submission_path)
else:
    train = add_features_to(trainingSet)
    features_to_include = ['Id', 'HelpfulnessNumerator', 'HelpfulnessDenominator', 'Time', 'Helpfulness']
    X_submission = pd.merge(testingSet, train[features_to_include], on='Id', how='left')

    # Drop redundant columns if they exist and rename columns if necessary
    if 'Score_x' in X_submission.columns:
        X_submission = X_submission.drop(columns=['Score_x'])
        X_submission = X_submission.rename(columns={'Score_y': 'Score'})

    # The training set where 'Score' is not null
    X_train = train[train['Score'].notnull()]

    X_submission.to_csv(submission_path, index=False)
    X_train.to_csv(train_path, index=False)
X_train, X_test, Y_train, Y_test = train_test_split(
    X_train.drop(columns=['Score']),
    X_train['Score'],
    test_size=1/4.0,
    random_state=0
)

import pandas as pd
from os.path import exists
from sklearn.model_selection import train_test_split

# Feature Extraction Function
def add_features_to(df):
    # Compute the helpfulness ratio safely, handling division by zero
    df['Helpfulness'] = df.apply(
        lambda row: row['HelpfulnessNumerator'] / row['HelpfulnessDenominator'] if row['HelpfulnessDenominator'] != 0 else 0,
        axis=1
    )
    return df

# Define paths
train_path = "/content/drive/My Drive/CS506/X_train.csv"
submission_path = "/content/drive/My Drive/CS506/X_submission.csv"

# Load pre-processed files if available
if exists(train_path) and exists(submission_path):
    X_train = pd.read_csv(train_path)
    X_submission = pd.read_csv(submission_path)
else:
    # Process the DataFrame and add features if pre-processed files do not exist
    train = add_features_to(trainingSet)
    features_to_include = ['Id', 'HelpfulnessNumerator', 'HelpfulnessDenominator', 'Time', 'Helpfulness']
    X_submission = testingSet.copy()
    for feature in features_to_include:
        if feature in train.columns and feature not in X_submission.columns:
            X_submission = pd.merge(X_submission, train[['Id', feature]], on='Id', how='left')

    # Drop redundant columns if they exist and rename columns if necessary
    if 'Score_x' in X_submission.columns:
        X_submission = X_submission.drop(columns=['Score_x'])
        X_submission = X_submission.rename(columns={'Score_y': 'Score'})

    # The training set where 'Score' is not null
    X_train = train[train['Score'].notnull()]

    # Save the feature-engineered files for future use
    X_submission.to_csv(submission_path, index=False)
    X_train.to_csv(train_path, index=False)

# Features to be used for training
features = ['HelpfulnessNumerator', 'HelpfulnessDenominator', 'Time', 'Helpfulness']

# Print columns to verify availability
print("X_train columns:", X_train.columns)
print("X_submission columns:", X_submission.columns)

# Select the feature columns from X_train and X_submission
if set(features).issubset(X_train.columns):
    X_train_select = X_train[features]
else:
    raise KeyError("Some of the required features are not available in X_train")

# Fill missing values in X_submission after merge to ensure the features are ready for predictions
for feature in features:
    if feature not in X_submission.columns:
        X_submission[feature] = 0  # Fill missing columns with a default value (e.g., 0)

X_submission[features] = X_submission[features].fillna(0)

if set(features).issubset(X_submission.columns):
    X_submission_select = X_submission[features]
else:
    raise KeyError("Some of the required features are not available in X_submission")

# Target variable for training
y = X_train['Score']

# Split the data into training and test sets (80% training, 20% testing)
X_train_split, X_test_split, y_train, y_test = train_test_split(X_train_select, y, test_size=0.2, random_state=42)

# Confirm the split shapes
print("X_train_split shape: ", X_train_split.shape)
print("X_test_split shape: ", X_test_split.shape)
print("y_train shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)

# Drop NaN values consistently from both X_train_select and y_train
combined_df = pd.concat([X_train_select, y_train], axis=1)

# Drop any rows where there are missing values
combined_df = combined_df.dropna()

# Split the data back into X_train_select and y_train
X_train_select = combined_df[features]
y_train = combined_df['Score']

# Confirm the shapes after dropping NaNs
print("After dropping NaNs:")
print("Shape of X_train_select:", X_train_select.shape)
print("Shape of y_train:", y_train.shape)
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3).fit(X_train_select, y_train)

# Predict the scores using the trained model on the test set
y_test_predictions = model.predict(X_test_split)

# Print predictions
print("Predicted values: ", y_test_predictions)

# Evaluate the model's performance
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(y_test, y_test_predictions)
conf_matrix = confusion_matrix(y_test, y_test_predictions)

# Print evaluation metrics
print("Accuracy: ", accuracy)
print("Confusion Matrix: \n", conf_matrix)

# Evaluate your model on the testing set
print("Accuracy on testing set = ", accuracy_score(y_test, y_test_predictions))

# Plot a confusion matrix
cm = confusion_matrix(y_test, y_test_predictions, normalize='true')
sns.heatmap(cm, annot=True)
plt.title('Confusion matrix of the classifier')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
import os

# Create the 'data' directory if it doesn't exist
if not os.path.exists('./data'):
    os.makedirs('./data')

# Create the submission file
X_submission['Score'] = model.predict(X_submission_select)
submission = X_submission[['Id', 'Score']]
submission.to_csv("./data/submission.csv", index=False)
```
